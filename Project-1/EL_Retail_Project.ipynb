```python
import pandas as pd

# Load the CSV
df = pd.read_csv("Superstore_Data.csv", encoding='ISO-8859-1')

# Convert both date columns to standard format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', format='%m/%d/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce', format='%m/%d/%Y')

# Drop rows where either date is invalid (NaT)
df = df.dropna(subset=['Order Date', 'Ship Date'])

# Optional: Convert dates back to string format YYYY-MM-DD for MySQL
df['Order Date'] = df['Order Date'].dt.strftime('%Y-%m-%d')
df['Ship Date'] = df['Ship Date'].dt.strftime('%Y-%m-%d')

# Save cleaned file
df.to_csv("Superstore_cleaned.csv", index=False)

print("✅ Cleaned file saved as Superstore_cleaned.csv")
```
